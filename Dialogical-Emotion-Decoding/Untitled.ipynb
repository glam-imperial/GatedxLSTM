```python

# import torch.nn as nn
# import numpy as np
# import pandas as pd
# from torch.utils.data import Dataset, DataLoader
# from collections import defaultdict
# from sklearn.metrics import classification_report, accuracy_score
# import numpy as np
# from sklearn.metrics import accuracy_score, f1_score
# from sklearn.model_selection import GroupShuffleSplit
# from torch.optim.lr_scheduler import ReduceLROnPlateau
# from torch.utils.data import DataLoader, random_split
# import torch
# import numpy as np
# import warnings
# from torch import nn, optim
# from torch.utils.data import DataLoader, Subset
# from sklearn.metrics import accuracy_score, f1_score, classification_report
# from sklearn.model_selection import StratifiedGroupKFold
# from collections import defaultdict
# from xlstm import (
#     xLSTMBlockStack,
#     xLSTMBlockStackConfig,
#     mLSTMBlockConfig,
#     mLSTMLayerConfig,
#     sLSTMBlockConfig,
#     sLSTMLayerConfig,
#     FeedForwardConfig,
# )
# import torch
# import random
# import numpy as np
# from sklearn.model_selection import train_test_split


# class Config:
#     # 数据参数
#     target_classes = {0: 0, 5: 1, 6: 2, 8: 3}
#     audio_dim = 1152   # 音频特征维度
#     text_dim = 1152     # 文本特征维度
#     feature_dim = audio_dim + text_dim  # 实际拼接后的维度 2304
    
#     # 模型参数
#     hidden_size = 256
#     num_layers = 2
#     batch_size = 16
#     num_epochs = 30

# class ContextualDataset(Dataset):
#     def __init__(self, metadata_path, audio_path, text_path):
#         metadata = pd.read_csv(metadata_path)
#         audio_data = np.load(audio_path, allow_pickle=True).item()
#         text_data = np.load(text_path, allow_pickle=True).item()
        
        
#         assert audio_data['features'].shape[1] == Config.audio_dim
#         assert text_data['features'].shape[1] == Config.text_dim
                
#         print(f"✅ Audio data shape: {audio_data['features'].shape}, Labels shape: {len(audio_data['labels'])}")
#         print(f"✅ Text data shape: {text_data['features'].shape}, Labels shape: {len(text_data['labels'])}")
#         # 检查前 10 个 label 是否匹配
#         for i in range(10):
#             print(f"🔍 Index {i} -> Audio Label: {audio_data['labels'][i]}, Text Label: {text_data['labels'][i]}")

#         valid_indices = [i for i, l in enumerate(audio_data['labels']) 
#                         if l in Config.target_classes]
        
#         self.features = np.concatenate([
#             audio_data['features'][valid_indices],
#             text_data['features'][valid_indices]
#         ], axis=1)
#         assert self.features.shape[1] == Config.feature_dim  # 确保2304维
    
#         self.labels = [Config.target_classes[l] 
#                       for l in audio_data['labels'][valid_indices]]
#         self.fileids = audio_data['fileid'][valid_indices]
        
#         self.fileid_to_idx = {fid:i for i,fid in enumerate(self.fileids)}
#         self.metadata = metadata[metadata['fileid'].isin(self.fileids)]
#         self.original_labels = audio_data['labels'][valid_indices]  # 新增行
#     def __len__(self):
#         return len(self.fileids)
    
#     def __getitem__(self, idx):
#         current_fid = self.fileids[idx]
#         current_features = self.features[idx]
        
#         # 上下文特征 (2+1+1)
#         context_features = []
#         for offset in [-2, -1, 0, 1, 2]:
#             try:
#                 if offset < 0:  
#                     fid = self._get_prev_fid(current_fid, abs(offset))
#                 elif offset > 0: 
#                     fid = self._get_next_fid(current_fid, offset)
#                 else:
#                     fid = current_fid
#                 context_features.append(self.features[self.fileid_to_idx[fid]])
#             except:
#                 context_features.append(current_features) 


#         return {
#             'features': torch.FloatTensor(np.stack(context_features)), 
#             'label': torch.LongTensor([self.labels[idx]])
#         }
    
#     def _get_prev_fid(self, base_fid, steps):
#         fid = base_fid
#         for _ in range(steps):
#             fid = self.metadata[self.metadata['fileid'] == fid]['prev'].values[0]
#         return fid
    
#     def _get_next_fid(self, base_fid, steps):
#         fid = base_fid
#         for _ in range(steps):
#             fid = self.metadata[self.metadata['fileid'] == fid]['next'].values[0]
#         return fid
    
    
# class ContextxLSTM(nn.Module):
#     def __init__(self):
#         super().__init__()
#         self.encoder = nn.Linear(Config.feature_dim, Config.hidden_size)
        
#         # xLSTM配置
#         self.xlstm = xLSTMBlockStack(
#             config=xLSTMBlockStackConfig(
#                 mlstm_block=mLSTMBlockConfig(
#                     mlstm=mLSTMLayerConfig(
#                         conv1d_kernel_size=3,
#                         qkv_proj_blocksize=4,
#                         num_heads=2
#                     )
#                 ),
#                 context_length=5,
#                 num_blocks=Config.num_layers,
#                 embedding_dim=Config.hidden_size
#             )
#         )
#         self.classifier = nn.Sequential(
#             nn.Dropout(0.3),
#             nn.Linear(Config.hidden_size, 4))
    
#     def forward(self, x):
#         # x: [batch, 5, 1152]
#         x = self.encoder(x)  # [batch, 5, hidden]
#         outputs = self.xlstm(x)  # [batch, 5, hidden]
#         return self.classifier(outputs[:, 2, :])  # 取中间时间步



# # ✅ 1. 加载完整数据集
# full_dataset = ContextualDataset(
#     "../IEMOCAP_extracted_emotions.csv",
#     "../IEMOCAP_audio_features_IAAN_context_1152.npy",
#     "../IEMOCAP_text_features_IAAN_context_1152.npy"
# )

# # ✅ 2. 提取数据
# fileids = np.array(full_dataset.fileids)  # (num_utterances,)
# features = np.array([sample["features"].numpy() for sample in full_dataset])  # (num_utterances, feature_dim)
# labels = np.array([sample["label"].item() for sample in full_dataset])  # (num_utterances,)

# # ✅ 3. 只保留目标类别 {0, 5, 6, 8}
# target_classes = {0: 0, 5: 1, 6: 2, 8: 3}
# filtered_fileids = []
# filtered_features = []
# filtered_labels = []

# for fid, feature, label in zip(fileids, features, labels):
#     if label in target_classes:
#         filtered_fileids.append(fid)
#         filtered_features.append(feature)
#         filtered_labels.append(target_classes[label])  # ✅ 确保只映射合法类别


# filtered_fileids = np.array(filtered_fileids)
# filtered_features = np.array(filtered_features)
# filtered_labels = np.array(filtered_labels)

# # ✅ 4. 生成 dialogs（对话结构）
# dialogs = defaultdict(list)
# for fid in filtered_fileids:
#     dialog_id = re.sub(r'_F\d+$|_M\d+$', '', fid)  # ✅ 去掉 `_Fxxx` 和 `_Mxxx`
#     dialogs[dialog_id].append(fid)

# print(f"✅ dialogs 结构: {type(dialogs)}, 长度: {len(dialogs)}")

# for key, value in list(dialogs.items())[:10]:  # 打印前 10 个对话
#     print(f"🟢 Dialog ID: {key}, Utterances: {value}")
    
# # 重点检查 `M` 说话人
# print("🔍 检查 `M` 说话人的 utterances:")
# for key, value in dialogs.items():
#     if "M" in key and len(value) == 0:
#         print(f"❌ 说话人 {key} 没有语句！")

# # ✅ 5. 生成 emo_dict（情感标签）
# emo_dict = {fid: label for fid, label in zip(filtered_fileids, filtered_labels)}

# print(f"✅ 原始数据量: {len(audio_data['labels'])}")
# print(f"✅ 过滤后数据量: {len(filtered_labels)}")

# # 检查前 10 个转换后的 labels
# for i in range(min(20, len(filtered_labels))):
#     print(f"🔍 FileID: {filtered_fileids[i]}, 原始 Label: {audio_data['labels'][i]}, 转换后: {filtered_labels[i]}")


# # ✅ 1. 加载训练好的 LSTM 模型
# model = ContextxLSTM()
# model.load_state_dict(torch.load("best_model.pth"))
# model.eval()  # 设为评估模式

# # ✅ 2. 生成 `out_dict`
# new_out_dict = {}

# with torch.no_grad():  # 关闭梯度计算
#     for idx in range(len(full_dataset)):  # 用索引遍历
#         fileid = full_dataset.fileids[idx]  # 获取 Utterance ID
#         features = full_dataset[idx]["features"].unsqueeze(0)  # 获取特征并增加 batch 维度
        
#         # 获取模型输出 (logits)，形状 [1, 4]
#         logits = model(features).squeeze().numpy()  # 转换为 numpy
        
#         new_out_dict[fileid] = logits  # 存入 DED 所需格式


# # print(dialogs, emo_dict,new_out_dict)


import torch.nn as nn
import numpy as np
import pandas as pd
from torch.utils.data import Dataset, DataLoader
from collections import defaultdict
from sklearn.metrics import classification_report, accuracy_score
import numpy as np
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import GroupShuffleSplit
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.utils.data import DataLoader, random_split
import torch
import numpy as np
import warnings
from torch import nn, optim
from torch.utils.data import DataLoader, Subset
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.model_selection import StratifiedGroupKFold
from collections import defaultdict
import re  # 新增导入
import random

# 假设xlstm模块已正确实现
from xlstm import (
    xLSTMBlockStack,
    xLSTMBlockStackConfig,
    mLSTMBlockConfig,
    mLSTMLayerConfig,
    sLSTMBlockConfig,
    sLSTMLayerConfig,
    FeedForwardConfig,
)

class Config:
    # 数据参数
    target_classes = {0: 0, 5: 1, 6: 2, 8: 3}
    audio_dim = 1152
    text_dim = 1152
    feature_dim = audio_dim + text_dim
    
    # 模型参数
    hidden_size = 256
    num_layers = 2
    batch_size = 16
    num_epochs = 30

class ContextualDataset(Dataset):
    def __init__(self, metadata_path, audio_path, text_path):
        metadata = pd.read_csv(metadata_path)
        audio_data = np.load(audio_path, allow_pickle=True).item()
        text_data = np.load(text_path, allow_pickle=True).item()
        
        # 保存原始标签用于验证
        self.original_audio_labels = audio_data['labels']
        self.original_text_labels = text_data['labels']
        
        # 验证特征维度
        assert audio_data['features'].shape[1] == Config.audio_dim
        assert text_data['features'].shape[1] == Config.text_dim
        
        # 获取有效索引（只在初始化时过滤一次）
        valid_indices = [
            i for i, l in enumerate(audio_data['labels']) 
            if l in Config.target_classes
        ]
        
        # 特征拼接
        self.features = np.concatenate([
            audio_data['features'][valid_indices],
            text_data['features'][valid_indices]
        ], axis=1)
        assert self.features.shape[1] == Config.feature_dim
        
        # 标签映射
        self.labels = [Config.target_classes[l] 
                      for l in audio_data['labels'][valid_indices]]
        self.fileids = audio_data['fileid'][valid_indices]
        self.original_labels = [l for l in audio_data['labels'][valid_indices]]  # 保存原始标签
        
        # 构建元数据映射
        self.fileid_to_idx = {fid:i for i,fid in enumerate(self.fileids)}
        self.metadata = metadata[metadata['fileid'].isin(self.fileids)].copy()
        
        # 在初始化数据集后添加
        print("\n=== 数据过滤验证 ===")
        print(f"原始音频标签数量: {len(full_dataset.original_audio_labels)}")
        print(f"过滤后有效样本数: {len(full_dataset)}")
        print(f"过滤掉样本数: {len(full_dataset.original_audio_labels) - len(full_dataset)}")

        # 验证标签映射
        unique_original = np.unique(full_dataset.original_labels)
        print(f"原始标签唯一值: {unique_original}")
        unique_mapped = np.unique(full_dataset.labels)
        print(f"映射后标签唯一值: {unique_mapped}")

        # 打印验证信息

    def __len__(self):
        return len(self.fileids)
    
    def __getitem__(self, idx):
        current_fid = self.fileids[idx]
        
        # 获取上下文特征（当前+前后各两个）
        context_features = []
        for offset in [-2, -1, 0, 1, 2]:
            try:
                if offset < 0:
                    fid = self._get_prev_fid(current_fid, abs(offset))
                elif offset > 0:
                    fid = self._get_next_fid(current_fid, offset)
                else:
                    fid = current_fid
                context_idx = self.fileid_to_idx[fid]
                context_features.append(self.features[context_idx])
            except KeyError:  # 处理边界情况
                context_features.append(self.features[idx])
        
        return {
            'features': torch.FloatTensor(np.stack(context_features)),
            'label': torch.LongTensor([self.labels[idx]])
        }
    
    def _get_prev_fid(self, base_fid, steps):
        current = base_fid
        for _ in range(steps):
            prev = self.metadata[self.metadata['fileid'] == current]['prev'].values
            if len(prev) == 0 or prev[0] == 'nan':
                return current  # 到达对话开头
            current = prev[0]
        return current
    
    def _get_next_fid(self, base_fid, steps):
        current = base_fid
        for _ in range(steps):
            next_ = self.metadata[self.metadata['fileid'] == current]['next'].values
            if len(next_) == 0 or next_[0] == 'nan':
                return current  # 到达对话结尾
            current = next_[0]
        return current

class ContextxLSTM(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Linear(Config.feature_dim, Config.hidden_size)
        
        self.xlstm = xLSTMBlockStack(
            config=xLSTMBlockStackConfig(
                mlstm_block=mLSTMBlockConfig(
                    mlstm=mLSTMLayerConfig(
                        conv1d_kernel_size=3,
                        qkv_proj_blocksize=4,
                        num_heads=2
                    )
                ),
                context_length=5,
                num_blocks=Config.num_layers,
                embedding_dim=Config.hidden_size
            )
        )
        self.classifier = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(Config.hidden_size, 4))
    
    def forward(self, x):
        x = self.encoder(x)  # [batch, 5, 2304] → [batch, 5, 256]
        outputs = self.xlstm(x)  # [batch, 5, 256]
        return self.classifier(outputs[:, 2, :])  # 取中间时间步

# ==================== 主流程 ====================
if __name__ == "__main__":
    # 初始化数据集
    full_dataset = ContextualDataset(
        "../IEMOCAP_extracted_emotions.csv",
        "../IEMOCAP_audio_features_IAAN_context_1152.npy",
        "../IEMOCAP_text_features_IAAN_context_1152.npy"
    )
    
    # 提取必要信息
    fileids = np.array(full_dataset.fileids)
    labels = np.array(full_dataset.labels)
    original_labels = np.array(full_dataset.original_labels)
    
    # 构建对话结构
    dialogs = defaultdict(list)
    for fid in fileids:
        dialog_id = re.sub(r'_F\d+$|_M\d+$', '', fid)
        dialogs[dialog_id].append(fid)
    
    # 构建标签字典
    emo_dict = {fid: label for fid, label in zip(fileids, labels)}
    
    # 加载模型
    model = ContextxLSTM()
    model.load_state_dict(torch.load("../best_model.pth"))
    model.eval()
    
    # 生成预测结果
    new_out_dict = {}
    with torch.no_grad():
        for idx in range(len(full_dataset)):
            fileid = fileids[idx]
            features = full_dataset[idx]["features"].unsqueeze(0)
            logits = model(features).squeeze().numpy()
            new_out_dict[fileid] = logits
    
    print("\n=== 键一致性验证 ===")
    missing_in_out = set(emo_dict.keys()) - set(out_dict.keys())
    missing_in_emo = set(out_dict.keys()) - set(emo_dict.keys())

    print(f"emo_dict 包含但 out_dict 缺失的键数: {len(missing_in_out)}")
    print(f"out_dict 包含但 emo_dict 缺失的键数: {len(missing_in_emo)}")

    # 示例输出
    if missing_in_out:
        print("示例缺失键:", list(missing_in_out)[:3])
    if missing_in_emo:
        print("示例多余键:", list(missing_in_emo)[:3])
```
